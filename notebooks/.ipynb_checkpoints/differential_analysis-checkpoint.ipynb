```r
import os
import pandas as pd

# count_featurecounts
count_dir = "./count_featurecounts"
output_matrix = "./count_featurecounts/gene_count_matrix_featurecounts.csv"


files = [f for f in os.listdir(count_dir) if f.endswith("_counts.txt")]
dfs = []

for file in files:
    sample_name = file.replace("_not_hg38_counts.txt", "")
    path = os.path.join(count_dir, file)
    
    # skiprows : ignorer les lligne commençant par #
    df = pd.read_csv(path, sep='\t', skiprows=1)
    
    # Garde uniquement 'Geneid' & dernière col
    df = df[['Geneid', df.columns[-1]]] 
    df.columns = ['Geneid', sample_name]
    dfs.append(df.set_index('Geneid'))


combined_df = pd.concat(dfs, axis=1)

# remove les gènes "unassigned"
combined_df = combined_df[~combined_df.index.str.contains("unassigned", na=True)]

# Sauvegarder la matrice finale
combined_df.to_csv(output_matrix)
print(f"Matrice : {output_matrix}")
```

```r
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Log transform
log_df = df.apply(lambda x: np.log(x + 1))

# PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(log_df.T)

# Plot 
plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_result[:,0], y=pca_result[:,1],  color='blue')
for i, txt in enumerate(df.columns):
    plt.annotate(txt, (pca_result[i,0], pca_result[i,1]),  fontsize=8)
plt.title("PCA - Échantillons RNA-seq")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()
# plt.figure(figsize=(10, 6))
# ax = sns.scatterplot(x=pca_result[:, 0], y=pca_result[:, 1], color='blue')

# for i, txt in enumerate(df.columns):
#     ax.text(pca_result[i, 0], pca_result[i, 1], txt, fontsize=8, ha='right')

# plt.title("PCA avec noms à gauche")
# plt.xlabel("PC1")
# plt.ylabel("PC2")
# plt.grid(True)
# plt.tight_layout()
# plt.show()
```

```r
library(DESeq2)

# Charger la matrice de comptage
countData <- read.csv("./count_featurecounts/gene_count_matrix_featurecounts.csv", row.names = "Geneid")

# Vérifier si tous les gènes sont à zéro
zero_genes <- apply(countData, 1, function(x) all(x == 0))
countData <- countData[!zero_genes, ]  # Enlever les gènes avec zéro partout

# Si plus aucun gène valide → sortir
if(nrow(countData) == 0) {
    stop("Tous les gènes sont à zéro. Pas de données valides pour DESeq2")
}

# Créer metadata manuellement (exemple avec 2 groupes)
metadata <- data.frame(
    sample = colnames(countData),
    group = c(rep("GroupA", 2), rep("GroupB", ncol(countData) - 2)),
    row.names = "sample"
)

# Créer le dataset DESeq2
dds <- DESeqDataSetFromMatrix(countData = countData,
                              colData = metadata,
                              design = ~ group)

# Filtrer les gènes avec au moins 10 lectures au total
keep <- rowSums(counts(dds)) >= 10
dds <- dds[keep, ]

# Lancer DESeq2
dds <- DESeq(dds)

# Obtenir les résultats
res <- results(dds)
write.csv(as.data.frame(res), "./DESeq2_results.csv")
```

```r
# test R:

# x <- c(1, 2, 3, 4, 5)
# mean_x <- mean(x)
# print(mean_x)
library(DESeq2)
library(ggplot2)

# Charger la matrice de comptage
countData <- read.csv("./count_featurecounts/gene_count_matrix_featurecounts.csv", row.names = "Geneid")

# Vérifier les noms des échantillons
print(colnames(countData))

# Définir les groupes manuellement (adapte cette partie selon tes échantillons)
# Adapte cette ligne si tu as plus ou moins de 4 échantillons :
num_samples <- length(colnames(countData))  # Nombre total d'échantillons

# Exemple : Tu définis GroupA et GroupB à partir des premiers échantillons
group_assignments <- c(rep("GroupA", 2), rep("GroupB", num_samples - 2))

# Créer le metadata avec tous les échantillons
metadata <- data.frame(
    sample = colnames(countData),
    group = group_assignments,
    row.names = "sample"
)

# Créer l'objet DESeq2
dds <- DESeqDataSetFromMatrix(countData = countData,
                              colData = metadata,
                              design = ~ group)

# Filtrer les gènes bien exprimés (ex: somme > 10 sur tous les échantillons)
keep <- rowSums(counts(dds)) >= 10
dds <- dds[keep, ]

# Lancer DESeq2
dds <- DESeq(dds)

# Obtenir les résultats
res <- results(dds)
res_df <- as.data.frame(res)

# Sauvegarder les résultats
write.csv(res_df, "./DESeq2_results.csv", row.names = TRUE)

# Volcano plot
res_df$significant <- ifelse(res_df$padj < 0.05 & abs(res_df$log2FoldChange) > 1,
                             yes = ifelse(res_df$log2FoldChange > 1, "Up", "Down"),
                             no = "Not Sig")

# Plot final
ggplot(res_df, aes(x = log2FoldChange, y = -log10(padj), color = significant)) +
    geom_point(alpha = 0.6, size = 2) +
    scale_color_manual(values = c("Up" = "red", "Down" = "blue", "Not Sig" = "gray")) +
    labs(title = "Volcano Plot - Gènes différentiellement exprimés",
         x = "log2(Fold Change)",
         y = "-log10(p-adjusted)") +
    theme_minimal() +
    geom_vline(xintercept = c(-1, 1), linetype = "dashed", color = "black") +
    geom_hline(yintercept = -log10(0.05), linetype = "dashed", color = "black") +
    theme(legend.position = "bottom")
```

```r
import matplotlib.pyplot as plt

# Charger les résultats DESeq2 depuis R
results = pd.read_csv("DESeq2_results.csv")

# Nettoyage
results = results[(results["padj"].notna()) & (results["log2FoldChange"].notna())]

# Créer une colonne pour significativité
results['significant'] = 'No'
results.loc[(results['padj'] < 0.05) & (results['log2FoldChange'] > 1), 'significant'] = 'Up'
results.loc[(results['padj'] < 0.05) & (results['log2FoldChange'] < -1), 'significant'] = 'Down'

# Plot
plt.figure(figsize=(10,6))
sns.scatterplot(data=results, x='log2FoldChange', y=-np.log10(results['padj']),
                hue='significant', palette={'Up': 'red', 'Down': 'blue', 'No': 'gray'},
                alpha=0.7, s=20)
plt.axhline(-np.log10(0.05), color='black', linestyle='--')
plt.axvline(1, color='black', linestyle='--')
plt.axvline(-1, color='black', linestyle='--')
plt.title("Volcano Plot : log2FC vs -log10(padj)")
plt.xlabel("log2(Fold Change)")
plt.ylabel("-log10(padj)")
plt.legend()
plt.show()
```
